```python
t=[]
num_threads=1
max_vert=30
for num_vert in range(5,max_vert,2):
    print(num_vert,end='')
    G = er(num_vert, 0.6*num_vert, True)
    tin=time.time()
    branch_and_bound(G,num_threads)
    t.append(time.time()-tin)
pl.plot(range(5,max_vert,2),t,'x-')
pl.title("Erdos Renyi weighted")
pl.xlabel("vertices")
pl.ylabel("solution time [sec]")
```
