```python
import numpy as np
import matplotlib.pyplot as pl
import time

import networkx as nx
import os
from matplotlib import rc
font = {'size' : 16}
rc('font', **font);
rc('text', usetex=True)

import sys
sys.path.append('../')


from classical_maxkcut_solver import *
from data.graph_generators import *
```

```python
num_threads=1
max_vert=27
for k in [2,3,4]:
    t=[]
    for num_vert in range(5,max_vert,2):
        print(num_vert,end='')
        G = er(num_vert, 0.6*num_vert, True)
        tin=time.time()
        classical_maxkcut_solver(G, k, num_threads)
        t.append(time.time()-tin)
    pl.semilogy(range(5,max_vert,2),t,'x-',label="k="+str(k))
pl.title("Erdos Renyi weighted")
pl.xlabel("vertices")
pl.ylabel("solution time [sec]")
pl.legend()
```
